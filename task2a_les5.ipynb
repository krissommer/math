```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import mlab as mlab
```

```python
k, m, a, b = 0, 0, 0, 0
n = 1000
for i in range(0, n):
    x = np.random.uniform(0, 10)
    y = np.random.uniform(0, 10)
    if x < 5:
        # print("первый орел")
        k = k + 1
    if y < 5:
        # print("второй орел")
        m = m + 1
    if x < 5 and y < 5:
        a = a + 1
    if x < 5 or y < 5:
        b = b + 1
        
print(0.5, 0.5, 0.25, 0.75)
print(k/n, m/n, a/n, b/n)
```
