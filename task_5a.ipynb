```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize = (5, 5))
ax = Axes3D(fig)
ax.set_title('Две_параллельные_плоскости')
# уравнения поверхностей: 3x+4y+5z-6=0 и 6x+8y+10z+12=0
f1 = lambda x, y: (6-3*x - 4*y)/5
f2 = lambda x, y: (-20-6*x - 8*y)/10
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)
z1 = f1(x, y)
z2 = f2(x, y)
surf = ax.plot_surface(x, y, z1, rstride=1, cstride=1, cmap=cm.plasma)
surf = ax.plot_surface(x, y, z2, rstride=1, cstride=1, cmap=cm.plasma)
show()
```
